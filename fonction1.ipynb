```python
import json 
import random
adresse_file = "culture_jazz_musiciens_2.json"
```

```python
def open_file (adresse_file):
    with open(adresse_file, 'r', encoding='utf-8') as archivo:
        data = json.load(archivo)
    return data
data = open_file (adresse_file)
```

```python
#choix du test
def test_utilisateur():
    print("=== Test culture JAZZ 1 ===")
    print("1. Nom de musicien -> Instrument")
    print("2. Nom de musicien -> Courant ")
    mode = int(input("\Choisisez le test (1 ou 2): "))
    print(f"Vous avez choisi le test numéro {mode}")
    return mode
i_test = test_utilisateur ()
```

```python
#fonction permettant de tester la connaissances des musiciens
def test_nom(data, i_test):
    m = random.choice(data)
    print(f"\nMusicien: {m['name']}")
    if (i_test == 1):
        correct_answers = m.get('instruments', [m.get('instrument')])
        answer = input("Donnez le nom d'un instrument joué par ce musicien : ").strip().lower()
    elif (i_test == 2):
        correct_answers = m.get('style', [m.get('style')])
        answer = input("Donnez le nom de la période à laquelle appartient ce musicien : ").strip().lower()
    else :
        print("Ce test nest pas envisagé")
        exit
    correct_answers_lower = [str(val).lower() for val in correct_answers]
    
    if answer in correct_answers_lower:
        print("✅ Votre réponse est correcte.")
    else:
        print("❌ Vous êtes nul! Voici la/les réponses attendues :")
        print(", ".join(correct_answers))
    return m

m=test_nom(data, i_test)
```

```python
#fonction permettant de tester la connaissances de musiciens jouant un instrument donné
def test_instrument(data):
    liste_instrument = sorted(list({inst for m in data for inst in m.get('instruments', []) if inst}))
    m_instrumento_objetivo = random.choice(liste_instrument).strip().lower()
    print(f"Instrument cible : {m_instrumento_objetivo}")
    
    answer = input("Donnez le nom d'un musicien qui joue cet instrument: ").strip().lower()

    if not answer:
        print("Veuillez entrer un nom.")
        return

    trouve = False
    for musicien in data:
        if answer in musicien['name'].lower():
            trouve = True
            sus_instruments = [i.lower() for i in musicien.get('instruments', [])]
            
            if m_instrumento_objetivo in sus_instruments:
                print(f"✅ Correct ! {musicien['name']} joue bien du {m_instrumento_objetivo}.")
            else:
                print(f"❌ {musicien['name']} est dans la base, mais ne joue pas de {m_instrumento_objetivo}.")
            break 

    if not trouve:
        print(f"❌ Désolé, aucun musicien avec le nom '{answer}' n'a été trouvé.")

test_instrument(data)
```
